## Saving extracted data
Once the data in python dictionary format, specific variables can be extracted and data will clean and ready for analysis. This approach will require some extra coding lines to ensure data integrity as shown in **method 1** below.

Another approach involves saving entire extracted data and subject it to further data cleaning process as shown in **method 2**

### method 1
Expracting specific varaiables.

```python
# "except" are added to handl missing values.
data=response.json()
results=[]

for p in data['products']:
    productName = p['productInfo']['productName']
    try:
        form = p['productInfo']['form']
    except: 
        form = 'na'
    unitPrice = p['productInfo']['unitPrice']
    size = p['productInfo']['productSize']
    productURL = p['productInfo']['productURL']
    productSize = p['productInfo']['productSize']
    averageRating = p['productInfo']['averageRating']
    try:
        reviewCount = p['productInfo']['reviewCount']
    except: 
        reviewCount = 'na'
        
    try:
        regularPrice = p['productInfo']['priceInfo']['regularPrice']
    except: 
        regularPrice = 'na'
    try:
        inactiveIngredients = p['productInfo']['inactiveIngredients']
    
    except:
        inactiveIngredients = 'na'

 
    vit = {'productName':productName,
           'form': form,
           'regularPrice': regularPrice,
           
           'unitPrice':unitPrice,
           'productURL':productURL,
           'productSize':productSize,
           'averageRating': averageRating,
           'reviewCount': reviewCount,
           'inactiveIngredients':inactiveIngredients  
          }

    

    results.append(vit)

import pandas as pd
df = pd.json_normalize(results)
df.to_csv("multivitamine_selected_columns.csv")
```

**Dealing with missing records in 'age' column**

```python
df3['age'].unique() # age variable is form of dictionary, the value needed to be extracted.
```

```python
# extracting the age value
p = "\:\s'(\w+\s?\w+)'"
df3['recommenedAge']=df3.age.str.extract(p,expand=True)
```

```python
df3['recommenedAge'].unique() # categories of "age"
```

```python
## Replacing "NaN" in "age" column with suitable sub-category 

df4 = df3.copy()
# converting  age to 'Adult' based on information in 'productName' column.
df4['recommenedAge2'] = np.where((df4.recommenedAge.isnull()) & 
                                 (df4.productName.str.contains('Wom')|
                                 df4.productName.str.contains('Man')|
                                 df4.productName.str.contains('Men')|
                                 df4.productName.str.contains('Her')|
                                 df4.productName.str.contains('Him')|
                                 df4.productName.str.contains('40')|
                                 df4.productName.str.contains('50')|
                                 df4.productName.str.contains('55')|
                                 df4.productName.str.contains('Tablet')|
                                 df4.productName.str.contains('Capsules')|
                                 df4.productName.str.contains('Adult')),'Adult', df4.recommenedAge)


# converting  age to 'Child' based on information in 'productName' column.
df4['recommenedAge3'] = np.where((df4.recommenedAge2.isnull()) & 
                                 (df4.productName.str.contains('Kid')|
                                 df4.productName.str.contains('Teen')|
                                 df4.productName.str.contains('Child')),'Child', df4.recommenedAge2)


# converting  age to 'Expectant Mother' based on information in 'productName' column.
df4['recommenedAge4'] = np.where((df4.recommenedAge3.isnull()) & 
                                 df4.productName.str.contains('Prenatal'),'Expectant Mother', df4.recommenedAge3)


# converting the remaining age records 'Others'based on information in 'productName' column.
df4['recommenedAge5'] = np.where((df4.recommenedAge4.isnull()),'Others', df4.recommenedAge4)
```

```python
df4.head(2)
```

```python
# remove the redundant "age" columns and rename the last transformed "age" column.
df5 = df4.drop(columns=['age', 'recommenedAge','recommenedAge2',
                  'recommenedAge3','recommenedAge4']).rename({'recommenedAge5': 'age'}, axis=1)


df5.head(2)
  
```

**Dealing with missing records in 'form' column**

```python
# Filling mising form base on information in 'productName' column
# replacing 'NaN' with 'Gummies'


df6 = df5.copy()
df6['form1'] = np.where((df6.form.isnull()) & df6.productName.str.contains('Gumm'),'Gummies',df6.form)


# Filling mising form base on information in 'productName' column
# replacing 'NaN' with 'Tablets'
df6['form2'] = np.where((df6.form1.isnull()) & df6.productName.str.contains('Tabl'),'Tablets',df6.form1)



# Replacing remaining 'NaN'  in 'form' column with " others" 
df6['form_final'] = np.where((df6.form2.isnull()),'Others',df6.form2)


```

```python
df6.head(2)
```

```python
# remove the redundant "form" columns and rename the last transformed "form" column.
df7 = df6.drop(columns=['form','form1','form2']).rename({'form_final': 'form'}, axis=1)


df7.head(2)
  
```

**Creating a binary variable column, whether the product contains Gelatin or not**

```python

df7['gelatin_content'] = np.where(df7.inactiveIngredients.str.contains('Gel'),'Yes','No')

df7.head(10)
#df7.gelatin_content.value_counts()
```
